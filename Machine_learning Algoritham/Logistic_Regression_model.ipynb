```python
for col_names in col_list:
  if data[col_names].dtypes == 'float64':
    sns.boxplot(data[col_names])
    plt.xlabel(col_names)
    plt.ylabel('Count')
    plt.show()
```

```python
# let us remove outliers first
```

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3-Q1
print(IQR)
```

```python
data = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
for col_names in col_list:
  if data[col_names].dtypes == 'float64':
    sns.boxplot(data[col_names])
    plt.xlabel(col_names)
    plt.ylabel('Count')
    plt.show()
```
