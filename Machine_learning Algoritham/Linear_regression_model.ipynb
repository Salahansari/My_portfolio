```python
for col_name in col:
  if data[col_name].dtypes == 'int64' or data[col_name].dtypes=='float64':
    sns.boxplot(data[col_name])
    plt.xlabel(col_name)
    plt.ylabel('Count')
    plt.show()
```

```python
# treating the outliers in every column
```

```python
for col_names in col:
  if data[col_names].dtypes == 'int64' or data[col_names].dtypes=='float64':
    Q3 = data[col_names].quantile(0.75)
    Q1 = data[col_names].quantile(0.25)
    IQR = Q3 - Q1
    data = data[(data[col_names] >= Q1 - 1.5*IQR) & (data[col_names]<= Q3 + 1.5*IQR)]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('charges',axis = 1)
Y = data['charges']
```

```python
X.columns
```

```python
x_train ,x_test,y_train , y_test = train_test_split(X , Y , train_size = 0.80 , random_state = 245)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
LR = LinearRegression()
```

```python
LR.fit(x_train , y_train)
```

```python
pred = LR.predict(x_test)
```

```python
from sklearn.metrics import *
```

```python
r2_test = r2_score(pred , y_test)
r2_test
```
